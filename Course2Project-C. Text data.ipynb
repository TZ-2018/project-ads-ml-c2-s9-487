C. Text data

You will filter and group your data based on the ingredients_text column. Your goal is
to find the five most common ingredients.

```python
#Importing and cleaning the data
#The full dataset has 356027 entries -> using full dataset, because filtered for one column
df = pd.read_csv('en.openfoodfacts.org.products.tsv',usecols=['ingredients_text'],nrows=356027,low_memory=False,sep='\t')
df.head(5)
```

```python
#Drop NaN
print('# NaN = ',df.isnull().sum().sum())
print('Before:',df.shape)
df2=df.dropna(axis=0)
print()
print('After:',df2.shape)
print(df2.head(5))
print()
print('# NaN = ',df2.isnull().sum().sum())
```

```python
#Split 'ingredients_text' into many columns
ingredients=df2.ingredients_text
ingredients[1]
```

```python
ingredients=ingredients.str.split(',', expand=True)
```

```python
ingredients
```

```python
#Make List and Drop NaN
df1=ingredients
print('# NaN = ',df1.isnull().sum().sum())
print(df1.shape)
s = df1.unstack()
df2=s.dropna(axis=0)
print()
print(df2.shape)
print(df2.head(5))
print()
print('# NaN = ',df2.isnull().sum().sum())
```

```python
# Count and find the five most common ingredients.
x=Counter(df2)
s=x.most_common()
print('The five most common ingredients:',s[0:5])
```
