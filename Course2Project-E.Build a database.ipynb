```python
import pandas as pd
import sqlite3
```

You will build a database to hold your data. It is up to you to defne appropriate tables
and well as primary keys for connecting them. (The focus is defnitely more on just
using the basic methods/tools introduced in the course and not on building a complex
database.) In particular, you can follow the following list of steps:

-restrict your data to 1000 entries and 5 columns of your choice

-create a connection to a sqlite3 database

-create one or multiple tables, at least one of the tables should have a PRIMARY KEY

-fill the database with your data

-run at least one query to demonstrate that it works correctly

```python
#Import Data
# -restrict your data to 1000 entries and 5 columns of your choice
#The full dataset has 356027 entries -> using subset 1000
df = pd.read_csv('en.openfoodfacts.org.products.tsv',
                 usecols=['code','creator','product_name','countries_en','carbohydrates_100g','proteins_100g','fat_100g'],
                 nrows=1000,low_memory=False,sep='\t')
```

```python
print(df.info())
print()
print("Datatypes:")
print(df.dtypes.head(9))
```

```python
# create a connection to a sqlite3 database
db = sqlite3.connect('Course2Project.db')
cursor= db.cursor()
```

```python
# create one or multiple tables, at least one of the tables should have a PRIMARY KEY
query="CREATE TABLE food (id INTEGER PRIMARY KEY, code INTEGER, creator TEXT, product_name TEXT, countries_en TEXT, fat_100g REAL, carbohydrates_100g REAL, proteins_100g REAL);"
cursor= db.cursor()
cursor.execute(query)
```

```python
# check that the table was created correctly by using a query we have already seen before
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
results = cursor.fetchall()
print(results)
```

```python
#fill the database with your data
df.to_sql(name='food', con=db, if_exists='append', index=False)
```

```python
# run at least one query to demonstrate that it works correctly
cursor.execute("SELECT * FROM food;")
results = cursor.fetchall()
#print(results)
results
```
